```python
import pandas as pd
import numpy as np
```

```python
#loading y_pred-predictions of events data  from different models:
y_te_pred_CB_Events=pd.read_csv('/content/drive/MyDrive/TalkingData/Events_Models/y_te_pred_CB.csv',index_col='device_id')
y_te_pred_LR_Events=pd.read_csv('/content/drive/MyDrive/TalkingData/Events_Models/y_te_pred_LR.csv',index_col='device_id')
y_te_pred_XG_Events=pd.read_csv('/content/drive/MyDrive/TalkingData/Events_Models/y_te_pred_XG.csv',index_col='device_id')
y_te_pred_NN_Events=pd.read_csv('/content/drive/MyDrive/TalkingData/Events_Models/y_te_pred_nn.csv',index_col='device_id')
```

```python
#loading y_pred-predictions of NO events data  from different models:
y_te_pred_CB_NoEvents=pd.read_csv('/content/drive/MyDrive/TalkingData/Noevents_Models/y_te_pred_CB.csv',index_col='device_id')
y_te_pred_LR_NoEvents=pd.read_csv('/content/drive/MyDrive/TalkingData/Noevents_Models/y_te_pred_LR.csv',index_col='device_id')
y_te_pred_XG_NoEvents=pd.read_csv('/content/drive/MyDrive/TalkingData/Noevents_Models/y_te_pred_XG.csv',index_col='device_id')
y_te_pred_NN_NoEvents=pd.read_csv('/content/drive/MyDrive/TalkingData/Noevents_Models/y_te_pred_nn.csv',index_col='device_id')
```

```python
y_pred_avg_Events=0.3*y_te_pred_CB_Events+0.7*y_te_pred_NN_Events
y_pred_avg_NoEvents=0.3*y_te_pred_CB_NoEvents+0.7*y_te_pred_NN_NoEvents
```

```python
y_test_pred=pd.concat([y_pred_avg_Events,y_pred_avg_NoEvents])
y_test_pred
```
